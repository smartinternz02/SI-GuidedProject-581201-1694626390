```python
sns.boxplot(df.YearsAtCompany)
```

```python
q1 = df.YearsAtCompany.quantile(0.25)
q3 = df.YearsAtCompany.quantile(0.75)
IQR = q3-q1
upper_limit = q3 + 1.5*IQR
```

```python
df['YearsAtCompany']= np.where(df['YearsAtCompany']>upper_limit,df.YearsAtCompany.median(),df['YearsAtCompany'])
```

```python
sns.boxplot(df.YearsAtCompany)
```

```python
sns.boxplot(df.MonthlyIncome)
```

```python
q1 = df.MonthlyIncome.quantile(0.25)
q3 = df.MonthlyIncome.quantile(0.75)
IQR = q3-q1
upper_limit = q3 + 1.5*IQR

df["MonthlyIncome"]= np.where(df['MonthlyIncome']>upper_limit,df.MonthlyIncome.median(),df['MonthlyIncome'])
sns.boxplot(df.MonthlyIncome)
```

```python
sns.boxplot(df.NumCompaniesWorked)
```

```python
q1 = df.NumCompaniesWorked.quantile(0.25)
q3 = df.NumCompaniesWorked.quantile(0.75)
IQR = q3-q1
upper_limit = q3 + 1.5*IQR

df["NumCompaniesWorked"]= np.where(df['NumCompaniesWorked']>upper_limit,df.NumCompaniesWorked.median(),df['NumCompaniesWorked'])
sns.boxplot(df.NumCompaniesWorked)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.Attrition = le.fit_transform(df.Attrition)
```

```python
df.BusinessTravel = le.fit_transform(df.BusinessTravel)
```

```python
df.Department = le.fit_transform(df.Department)
```

```python
df.Gender = le.fit_transform(df.Gender)
```

```python
df.MaritalStatus = le.fit_transform(df.MaritalStatus)
```

```python
df.EducationField = le.fit_transform(df.EducationField)
```

```python
df.JobRole = le.fit_transform(df.JobRole)
```

```python
df.Over18 = le.fit_transform(df.Over18)
```

```python
df.OverTime = le.fit_transform(df.OverTime)
```

```python
df.head()
```

```python
y = df['Attrition']
x = df.drop(['Attrition','EmployeeNumber','EmployeeCount','StockOptionLevel','Over18','StandardHours'], axis = 1)
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_scaled=pd.DataFrame(ms.fit_transform(x),columns=x.columns)
```

```python
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.head()
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
x.head()
```

```python
x.iloc[1].values
```

```python
model.predict(ms.transform([x.iloc[1].values]))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
probability=model.predict_proba(x_test)[:,1]
```

```python
probability
```

```python
fpr,tpr,threshsholds = roc_curve(y_test,probability)
```

```python
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={
 'criterion':['gini','entropy'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']

}
```

grid_search.fit(x_train,y_train)

```python
dtc_cv=DecisionTreeClassifier(criterion= 'entropy',
 max_depth=3,
 max_features='sqrt',
 splitter='best')
dtc_cv.fit(x_train,y_train)
```

## Random forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
forest_params = [{'max_depth': list(range(10, 15)), 'max_features': list(range(0,14))}]
```

```python
rfc_cv= GridSearchCV(rfc,param_grid=forest_params,cv=10,scoring="accuracy")
```

```python
rfc_cv.fit(x_train,y_train)
```

```python
pred=rfc_cv.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```
